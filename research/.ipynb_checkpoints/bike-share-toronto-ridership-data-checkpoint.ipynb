# Station API

```python
base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"
url = base_url + "/api/3/action/package_show"
```

```python
params = { "id": "bike-share-toronto"}
package = requests.get(url, params=params).json()
pprint(package)
```

```python
bike_share_json = requests.get(package["result"]["resources"][0]['url']).json()
pprint(bike_share_json)
```

```python
# Access the json data with the name 'station_information'
station_information_json = requests.get(bike_share_json['data']['en']['feeds'][2]['url']).json()
pprint(station_information_json)
```

```python
# How many bike stations are there in total?
station_id_todate = [int(station['station_id']) for station in station_information_json['data']['stations']]
print(f'There are currently {len(station_id_todate)} stations in the system.')
```

```python
df.info(show_counts=True)
```

- Each trip is marked by a unqiue `Trip Id`, which means that they can be simply dropped.
- No missing entries for `Bike Id` or `User Type`. There are only 2 types of values for `User Type`, Annual and Casual.
- The dataset needs to go through a few passes of cleaning.
- Other than data type conversions, first pass needs to be done based on `Trip Duration`. We need to eliminate obvious outliers, such as those with 0 sec duration.
- Second pass is for when we want to focus on geospatial related statistics.
- Start time and end time should be converted to data-time object for easier data manipulation downstream.


## Stage 1: Data Type Conversions

```python
df_1 = df.copy()
df_1.drop(columns=['Trip Id'], inplace=True)
```

```python
df_1['Start Time'] = pd.to_datetime(df_1['Start Time'])
df_1['End Time'] = pd.to_datetime(df_1['End Time'])
```

Very interestingly, based on the results shown below, the majority of null entries for `End Station Id` have a trip duration of 0 sec, so we will circle back to this after we deal with the outliers in `Trip Duration`. 

```python
df_1[df_1['End Station Id'].isnull()]
```

```python
df_1['End Station Id'] = df_1['End Station Id'].astype('Int64')
```

```python
# Reorder the columns
df_1 = df_1.iloc[:, [2,5,0,1,3,4,6,7,8]]
```

```python
df_1.info(show_counts=True)
```

## Stage 2: Outliers in Trip Duration

```python
df_2 = df_1.copy()
df_2.rename(columns={'Trip  Duration': 'Duration'}, inplace=True)

# Remove trip duration of 0.
df_2 = df_2[df_2['Duration']>0]
len(df_2['Duration']) # This operation alone eliminates 4620469 - 4615878 = 4591 records.
```

```python
# Check if there is any trip less than a min. Will decide whether to include this based on a well-reasoned cutoff point.
len(df_2[df_2['Duration']<60])
```

```python
# Compute the typical outlier detectors.
q1 = df_2['Duration'].quantile(0.25)
q3 = df_2['Duration'].quantile(0.75)
iqr = q3 - q1
print(q1, q3, iqr)
```

```python
outliers = (df_2['Duration']<(q1-1.5*iqr)) | (df_2['Duration']>(q3+1.5*iqr))
# Note that all entries with positive durations are kept.
len(df_2[~outliers]['Duration']) # 4615878 - 4454449 = 161429 records.
```

Interestingly, only 13 records without End Station Id left.

```python
df_2[~outliers][df_2[~outliers]['End Station Id'].isnull()]
```

```python
df_2[~outliers]['Duration'].hist(bins=30)
plt.xlim([0, q3+1.5*iqr])
```

```python
df_2[~outliers][df_2[~outliers]['User Type'] == 'Annual Member']['Duration'].hist(bins=30)
plt.xlim([0, q3+1.5*iqr])
```

```python
df_2[~outliers][df_2[~outliers]['User Type'] == 'Casual Member']['Duration'].hist(bins=30)
plt.xlim([0, q3+1.5*iqr])
```

The way the distribution for casual members trails off here looks like information lost more than outlier removal. As a result, the quantiles are determined and the outliers removed for each group.

```python
# Compute the outliers for annual members.
df_2_annual = df_2[df_2['User Type'] == 'Annual Member']
q1 = df_2_annual['Duration'].quantile(0.25)
q3 = df_2_annual['Duration'].quantile(0.75)
iqr = q3 - q1
print(q1, q3, iqr)

df_2_annual = df_2_annual[~((df_2_annual['Duration']<(q1-1.5*iqr)) | (df_2_annual['Duration']>(q3+1.5*iqr)))]
df_2_annual['Duration'].hist(bins=30)
plt.xlim([0, q3+1.5*iqr])
```

```python
# Compute the outliers for casuals.
df_2_casual = df_2[df_2['User Type'] == 'Casual Member']
q1 = df_2_casual['Duration'].quantile(0.25)
q3 = df_2_casual['Duration'].quantile(0.75)
iqr = q3 - q1
print(q1, q3, iqr)

df_2_casual = df_2_casual[~((df_2_casual['Duration']<(q1-1.5*iqr)) | (df_2_casual['Duration']>(q3+1.5*iqr)))]
df_2_casual['Duration'].hist(bins=30)
plt.xlim([0, q3+1.5*iqr])
```

```python
pd.concat([df_2_annual, df_2_casual]) # Can assign this to df_2 if this approach is used.
```

```python
# For now
df_2 = df_2[~outliers]
df_2.info(show_counts=True)
```
